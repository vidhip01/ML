# Correlation with Heatmap

Tip is dependent feature
Total Bill, Sex, SMoker, Day etc are independent feature

A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix (table) between two discrete dimensions or event types. It is very important in Feature Selection

```python
df['smoker'] = df['smoker'].map({'yes': 1, 'no': 0})
```

```python
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns
numerical_df = df[numerical_columns]
```

# Join Plot
A join plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='hex')
```

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='reg')
#reg is regression
#it draws TDF
```

# Pair Plot (for more than two independent features)
A “pairs plot” is also known as a scatterplot, in which one variable in the same data row is matched with another variable's value, like this: Pairs plots are just elaborations on this, showing all variables paired with all the other variables

```python
sns.pairplot(df)

#doing permutation and combination on all values
```

```python
sns.pairplot(df, hue='sex')
```

```python
sns.pairplot(df, hue='smoker')
```

```python
df['smoker'].value_counts()
```

# Dist Plot
helps to check the distribution of the columns features

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'], kde=False, bins=10)
```

kde parameter stands for kernel density estimate. When kde is set to True (the default), it adds a smoothed curve to the histogram, providing an estimate of the probability density function (PDF) of the data.

# Categorical Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='sex', data=df)
plt.show()
```

# Box Plot

A box and whisker plot(sometimes called boxplot) is a graph that presents information
from a five-number summary

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x='day', y='total_bill', data=df, palette='rainbow')
plt.show()
```

this is in percentile
the dots outside are outliers

```python
sns.boxplot(x='sex', y='total_bill', data=df, palette='rainbow')
plt.show()
```

```python
sns.boxplot(data=df, orient='v')
```

Box plot doesn't need to have only categorical features

```python
sns.boxplot(x="total_bill", y="day", hue="smoker", data=df)
```
